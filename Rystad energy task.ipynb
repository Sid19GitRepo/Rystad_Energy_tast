# Task 1 data gathering

```python
#useful libraries for scraping the web
from bs4 import BeautifulSoup as soup
import requests
from urllib.request import urlopen as urlr
import pandas as pd  # for the data processing 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#accessing and reading the html format of the web https://www.eia.gov/dnav/pet/pet_pnp_inpt_a_epc0_yir_mbbl_m.htm
web_url = 'https://www.eia.gov/dnav/pet/pet_pnp_inpt_a_epc0_yir_mbblpd_m.htm'
web_html = urlr(web_url)
page_html = web_html.read()
page_soup = soup(page_html, 'html.parser')
tables = page_soup.findAll('table', {'class': 'data1'})
print(type(tables))
```

```python
#finding the link of content insoecting the web page
link_ = page_soup.a
link_xlsx = link_.get('href').replace('./','')
print(link_xlsx)
```

```python
#download link
link = web_url.replace('pet_pnp_inpt_a_epc0_yir_mbblpd_m.htm', link_xlsx) # attaching the download link with server page
r = requests.get(link) 
# downloading the data as excel worksheet
with open('data.xlsx', 'wb') as f:  
    f.write(r.content)
f.close()   
# downloading the data as csv
with open('data.csv', 'wb') as f:  
    f.write(r.content)
f.close()
```

```python
# a. reading the time series of U.S. Refinery and Blender Net Input of Crude Oil
df=pd.read_excel('data.xlsx','Data 1',header=2)
df.tail(2)
sns.plot(x='Date',y='U.S. Refinery and Blender Net Input of Crude Oil (Thousand Barrels per Day)', data=df)
```

# Task 2. Data processing 

```python
# b. keeping data onwards 2016
series=[]
for i in range(len(df)):
    if df.Date[i].year>2015:
        series.append(True)
    else:   
        series.append(False)
```

```python
required_df = df[series]
row_df=df.Date[series]
new_df=pd.DataFrame(row_df, columns=['Date']) # applying series on date column
required_df.head()
```

```python
# c. creating year column
new_df['year']=[new_df.Date[i].year for i in range(733-len(new_df),733)]

# c. creating quater column
list1=[]
for i in range(733-len(new_df),733):
    if (new_df.Date[i].month/3)<=1:
        list1.append('quarter_1')
    elif (new_df.Date[i].month/3>1) and (new_df.Date[i].month/3<=2):
        list1.append('quarter_2')
    elif (new_df.Date[i].month/3>2) and (new_df.Date[i].month/3<=3):
        list1.append('quarter_3')
    elif (new_df.Date[i].month/3>3) and (new_df.Date[i].month/3<=4):
        list1.append('quarter_4')    
new_df['qurter']=list1

# c. creating month column
new_df['month']=[new_df.Date[i].month for i in range(733-len(new_df),733)]

# c. creating columns for PADD1, PADD2, PADD3, PADD4, PADD5
new_df['(PADD 1) Refinery and Blender Net Input of Crude Oil'] = required_df['East Coast (PADD 1) Refinery and Blender Net Input of Crude Oil (Thousand Barrels per Day)']
new_df['(PADD 2) Refinery and Blender Net Input of Crude Oil'] = required_df['Midwest (PADD 2) Refinery and Blender Net Input of Crude Oil (Thousand Barrels per Day)']
new_df['(PADD 3) Refinery and Blender Net Input of Crude Oil'] = required_df['Gulf Coast (PADD 3) Refinery and Blender Net Input of Crude Oil (Thousand Barrels per Day)']
new_df['(PADD 4) Refinery and Blender Net Input of Crude Oil'] = required_df['Rocky Mountain (PADD 4) Refinery and Blender Net Input of Crude Oil (Thousand Barrels per Day)']
new_df['(PADD 5) Refinery and Blender Net Input of Crude Oil'] = required_df['West Coast (PADD 5) Refinery and Blender Net Input of Crude Oil (Thousand Barrels per Day)']
```

```python
# d. column for Total US Refinery Net Input of Crude Oil
new_df['Total US Refinery Net Input of Crude Oil'] = new_df.iloc[:, -5:-1].sum(axis=1)
new_df = new_df.reset_index(drop=True)
new_df.head(12)
```

```python
# e. Quaterly report
quarter_df_list=[]
for i in range(0,len(new_df),3):
    q_df=new_df.iloc[i:i+4, -6:].sum(axis=0)
    q_df_ =pd.DataFrame(q_df)
    q_df_=q_df_.swapaxes("index", "columns")
    q_df_['year_quarter']=str(new_df['year'][i]) +' '+ new_df['qurter'][i]
    q_df_.set_index(['year_quarter'], inplace = True)
    quarter_df_list.append(q_df_/4)
result_quarter = pd.concat(quarter_df_list)    
result_quarter.to_csv('quarter_report.csv') # data as csv
result_quarter.to_excel('quarter_report.xlsx') # data as excel worksheet
```

```python
# f. Yearly report
year_df_list=[]
for i in range(0,len(new_df),12):
    y_df1=new_df.iloc[i:i+12, -6:].sum(axis=0)
    y_df_1 =pd.DataFrame(y_df1)
    y_df_1=y_df_1.swapaxes("index", "columns")
    y_df_1['year']=new_df['year'][i]
    y_df_1.set_index(['year'], inplace = True)
    year_df_list.append(y_df_1/12)
    #print(y_df_1)
result_year = pd.concat(year_df_list)    
result_year
result_year.to_csv('yearly_report.csv') # data as csv
result_year.to_excel('yearly_report.xlsx') # data as excel worksheet
```
